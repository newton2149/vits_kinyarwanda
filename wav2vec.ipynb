```python


import torch
import torchaudio
from transformers import AutoModel
import os

# Load the Wav2Vec2 model.
wav2vec2_model = AutoModel.from_pretrained("facebook/wav2vec2-base")
wav_dir = 'cv-corpus-5.1-2020-06-22/rw/selected_clips'

# Extract speaker embeddings from the audio recordings.
```

```python
speaker_embeddings = []
for filename in os.listdir(wav_dir):
    audio_tensor, sample_rate = torchaudio.load(os.path.join(wav_dir, filename))
    # audio_tensor = torch.from_numpy(audio_waveform)
    speaker_embedding = wav2vec2_model(audio_tensor).last_hidden_state[:, 0].detach().cpu().numpy()
    speaker_embeddings.append(speaker_embedding)

```

```python
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

def compute_cosine_similarity_with_sklearn(embedding1, embedding2):
  """Computes the cosine similarity between two embeddings using sklearn.

  Args:
    embedding1: A NumPy array of the first embedding.
    embedding2: A NumPy array of the second embedding.

  Returns:
    A float representing the cosine similarity between the two embeddings.
  """

  cosine_similarity = cosine_similarity(embedding1.reshape(1, -1), embedding2.reshape(1, -1))[0][0]

  return cosine_similarity
```
